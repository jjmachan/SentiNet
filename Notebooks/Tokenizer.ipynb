```python
import os
import pandas as pd
from collections import Counter
import pickle
```

```python
import spacy
nlp = spacy.load('en_core_web_sm', disable=['tagger', 'ner', 'parser'])
```

```python
data = '../Data/'

train_pos_filename = os.listdir(data)

columns = ['Index', 'Sentence', 'Rating', 'isPositive', 'SentVec']
df = pd.read_table(data+'/train.csv', names=columns)
df
```

### loading the dictionary

If the dictionaries are saved execute this and skip the next section

```python
print(train_pos_filename)

word2idx = pickle.load(open(f'{data}/word2idx.pkl', 'rb'))
idx2word = pickle.load(open(f'{data}/idx2word.pkl', 'rb'))
```

### Create the dictionaries for conversion

```python
#creates a new counter and populates it
words = Counter()

for review in df.Sentence:
    words.update(token.text.lower() for token in nlp(review))
    #print(review)
```

```python
words
```

```python
#Sort the string and add 2 new tokens
words = sorted(words, key=words.get, reverse=True)
words = ['_PAD','_UNK'] + words
```

```python
len(words)
```

```python
# create word to index dictionary and reverse
word2idx = {o:i for i,o in enumerate(words)}
idx2word = {i:o for i,o in enumerate(words)}
```

```python
idx2word
```

```python
pickle.dump(idx2word, open(f'{data}/idx2word.pkl', 'wb'))
pickle.dump(word2idx, open(f'{data}/word2idx.pkl', 'wb'))
```

```python
vector = []
for s in df.Sentence:
    vector.append([word2idx[w.text.lower()] for w in nlp(s)])
```

```python
df.SentVec = vector
```
