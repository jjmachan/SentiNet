```python
import re

REPLACE_NO_SPACE = re.compile("(\.)|(\;)|(\:)|(\!)|(\')|(\?)|(\,)|(\")|(\()|(\))|(\[)|(\])")
REPLACE_WITH_SPACE = re.compile("(<br\s*/><br\s*/>)|(\-)|(\/)")

def preprocess_reviews(line):
    line = REPLACE_NO_SPACE.sub("", line)
    line = REPLACE_WITH_SPACE.sub(" ", line)
    
    return line
```

### Generating the dataframes

This step generated the required dataframe for the test and train files. First run the loop for '/train' and then for '/test'.

```python
for i in ['/test']:
    for j in ['/pos/', '/neg/']:
        if j == '/pos/':
            isPositive = 1
        else:
            isPositive = 0
        #print(os.listdir(data+i+j))
        fileNames = os.listdir(data+i+j)
        for file in fileNames:
            #print(file[:-4].split('_'))
            line = ''
            idx, rating = file[:-4].split('_')
            with open(data+i+j+file) as f:
                #print(f.readline())
                line = f.readline()
            line = preprocess_reviews(line)
            row = {
                    'Index': idx,
                    'Sentence': line,
                    'Rating': rating,
                    'isPositive': isPositive
                  }
            #print(row)
            df = df.append(row, ignore_index=True)
            
```
